```python
# Get Topic Prominence TS

topic_prominences = []
for c,(i,r) in enumerate(cab_.iterrows()):
    mask = (dists.index > r['startdate']) & (dists.index <= r['enddate'])
    d_ = dists.loc[mask]
    if len(d_) > 10:
        d_ = d_.mean().reset_index()
        d_.columns = ['topic','p']
        d_['cab'] = r['cabinet']
        topic_prominences.append(d_)
topic_prominences = pd.concat(topic_prominences).pivot(index='cab',columns='topic',values='p')
```

```python
topic_prominences = topic_prominences.to_dict('records')
```
